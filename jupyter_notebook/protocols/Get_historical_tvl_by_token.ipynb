```python
response = requests.get('https://api.llama.fi/protocols')
data = response.json()
df = pd.DataFrame.from_dict(data, orient = 'Columns')
names = df.slug
```

```python
failed = 0
count = 0
timestamp = datetime.now()

try:
    with connect_db('protocols_historic_tvl_by_tokens') as connection:
        for coin in names:
            try:
                response = requests.get(f'https://api.llama.fi/protocol/{coin}')
                response.raise_for_status()  # Raise an HTTPError for bad responses
                data = response.json()
                tokens = data['tokens']
                table_name = coin

                dataList = []

                for daily_data in tokens:
                    date = daily_data['date']
                    token_data = daily_data.get('tokens', {})

                    for coin, total_liquidity in token_data.items():
                        if total_liquidity is not None:
                            dataList.append({'date': date, 'tvl': total_liquidity, 'protocol': coin})

                df = pd.DataFrame(dataList)
                if not df.empty:
                    df['date'] = pd.to_datetime(df['date'], unit='s')
                    df['updated_at'] = timestamp

                    # Write the DataFrame to a table named after the coin
                    df.to_sql(table_name, con=connection, if_exists='replace', index=False)

                    count += 1
                    print(f'Processed data for {table_name}! - {len(names) - count} to go')

            except Exception as e:
                print(f'Error processing {coin}: {e}')
                failed += 1

except Exception as e:
    print(f'Error connecting to the database: {e}')
    failed += 1

print(f'Number of failures: {failed}')
```
