## Historic Price of Coins (top 1250) 


```python
with connect_db('coins') as connection:
    count = 0 
    failed  =0
    coins_failed = []
    timestamp = datetime.now()
    for coin in names:
        try:
            response = requests.get(f'https://api.coingecko.com/api/v3/coins/{coin}/ohlc?vs_currency=usd&days=max')

            # Check if the response status code indicates success
            if response.status_code == 200:
                try:
                    data = response.json()
                    df = pd.DataFrame.from_dict(data)
                    df = df.rename(columns={
                        0: 'date',
                        1: 'open',
                        2: 'high',
                        3: 'low',
                        4: 'close'
                    })
                    df['date'] = pd.to_datetime(df['date'], unit='ms')
                    df['id'] = coin
                    # Add Simple Moving Averages (SMA) and Exponential Moving Averages (EMA)
                    df['sma_10'] = df['close'].rolling(window=10).mean()
                    df['sma_25'] = df['close'].rolling(window=25).mean()
                    df['sma_50'] = df['close'].rolling(window=50).mean()
                    df['sma_100'] = df['close'].rolling(window=100).mean()

                    df['ema_10'] = df['close'].ewm(span=10, adjust=False).mean()
                    df['ema_25'] = df['close'].ewm(span=25, adjust=False).mean()
                    df['ema_50'] = df['close'].ewm(span=50, adjust=False).mean()
                    df['ema_100'] = df['close'].ewm(span=100, adjust=False).mean()
                    
                    # Calculate Daily Returns and Log Returns
                    df['daily_return'] = df['close'].pct_change()
                    df['log_return'] = np.log(1 + df['daily_return'])

                    # Calculate Volatility
                    df['volatility'] = df['log_return'].rolling(window=20).std()

                    # Calculate Relative Strength Index (RSI)
                    delta = df['close'].diff()
                    gain = delta.where(delta > 0, 0)
                    loss = -delta.where(delta < 0, 0)
                    avg_gain = gain.rolling(window=14).mean()
                    avg_loss = loss.rolling(window=14).mean()
                    rs = avg_gain / avg_loss
                    df['rsi'] = 100 - (100 / (1 + rs))

                    # Calculate Moving Average Convergence Divergence (MACD)
                    df['ema_12'] = df['close'].ewm(span=12, adjust=False).mean()
                    df['ema_26'] = df['close'].ewm(span=26, adjust=False).mean()
                    df['macd'] = df['ema_12'] - df['ema_26']
                    df['signal_line'] = df['macd'].ewm(span=9, adjust=False).mean()
                    
                    df['updated_at'] = timestamp
                    table_name = coin
                    df.to_sql(table_name, con=connection, if_exists='replace')
                    time.sleep(6)
                    count += 1
                    print(f'{coin} inserted - {len(names) - count} to go')

                except ValueError as e:
                    print(f"Error processing data for ID {coin}: {e}")
                    failed += 1
                    coins_failed.append(coin)
            else:
                print(f"API request failed for coin {coin}. Status code: {response.status_code}")
                failed += 1    
                coins_failed.append(coin)
                

        except requests.exceptions.RequestException as e:
            print(f"Error making API request for coin {coin}: {e}")
            failed += 1
            coins_failed.append(coin)
            
    print (failed)
```
